Importing The Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Reading and Explore The Data

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
df.head()
```

```python
df = df.rename(columns = {"Flight Distance": "Flight_Distance", "Departure Delay": "Departure_Delay_in_Minutes", 'Arrival Delay': 'Arrival_Delay_in_Minutes'})
```

```python
df.isna().sum()
```

```python
df.duplicated().any()
print(df.columns)
```

```python
numircal_cols = ['ID','Age','Flight_Distance','Departure_Delay_in_Minutes','Arrival_Delay_in_Minutes']
vars_categorical = ['Gender', 'Customer Type', 'Ease of Online Booking', 'Check-in Service', 'Online Boarding',
       'Gate Location', 'On-board Service', 'Seat Comfort', 'Leg Room Service',
       'Cleanliness', 'Food and Drink', 'In-flight Service',
       'In-flight Wifi Service', 'In-flight Entertainment', 'Baggage Handling',
       'Satisfaction']
```

Getting summary statistics on the data

```python
stats=df[numircal_cols].describe().T.rename_axis("Col_Name").reset_index()
stats
```

Checking For Outliers

```python
for col in numircal_cols:
    plt.figure(figsize = (10,2))
    sns.boxplot(data=df,x=col)
    
```

There Is Outliers (Flight Distance,Arrival Delay,Departure Delay)

```python
df = df.dropna()
df.isnull().any().sum()
```

Visualization

```python
x=df["Satisfaction"].value_counts()
```

```python
insights=round(df['Satisfaction'].value_counts(normalize=True)*100,2)
plt.pie(insights.values,labels=insights.index,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'},shadow=True,startangle=90)
plt.title("Distribution of Satisfaction for Airline Passengers")
plt.show()
```

Most of them Not Satisfied

(Gender,Customer,Type)

```python
cols=['Gender','Customer Type','Type of Travel']
colors = sns.color_palette('pastel')[0:5]
for col in cols:
    x=round(df[col].value_counts(normalize=True)*100,2)
    plt.pie(x.values,labels=x.index,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'},shadow=True,startangle=90,colors=colors)
    plt.title('Distribution for Passengers that has ' + col , weight = 'bold')
    plt.show()
    
```

Female Passangers are More Than Male
Most of The passangers Are Bussiness Class And Not their First Time

Getting The relation Between Satisfaction And Numircal Values

```python
for col in numircal_cols:
    sns.barplot(x = 'Satisfaction', y = col, data = df, errorbar=None)
    plt.title('Relation between Passenger_Satisfaction and ' + col)
    plt.show()
```

Relation Between Categorical and Satisfication

```python
Cat_cols=[n for n in vars_categorical if n!="Satisfaction"]
for cols in Cat_cols:
    
    
    sns.countplot(data = df, 
                  x = 'Satisfaction', 
                  hue = cols)
    
    plt.title('Distribution for Passenger Satisfaction for has ' + cols)
    plt.show()
                   
                
     
```
